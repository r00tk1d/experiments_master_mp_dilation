```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tssb.utils import load_time_series_segmentation_datasets

# data 
df_ts = load_time_series_segmentation_datasets()
window_ranges = df_ts['window_size'].tolist()
number_of_cps = [len(x)+1 for x in df_ts['change_points'].tolist()]
ts_lengths = [len(x) for x in df_ts['time_series'].tolist()]
```

## Load Data ##

```python

df_scores_known_target_w = pd.read_csv("../_segmentation_covering_known_stumpy_target_w_fluss_ALL.csv")
df_scores_known_m = pd.read_csv("../_segmentation_covering_known_stumpy_m_fluss_ALL.csv")
df_scores_unknown_target_w = pd.read_csv("../_segmentation_covering_unknown_stumpy_target_w_fluss_ALL.csv")
df_scores_unknown_m = pd.read_csv("../_segmentation_covering_unknown_stumpy_m_fluss_ALL.csv")
```

## Visualize Correlations fluss1 vs. flussOracle: known, fixed target w ##

```python
fontsize = 18
experiment_name = 'correlation_known_fixed_target_w_'

# Pearson Correlations
print(f'Pearson Correlation without dilation, target w: {np.corrcoef(window_ranges,  df_scores_known_target_w["fluss1"].tolist())[1,0]}')
print(f'Pearson Correlation with dilation (Oracle), target w: {np.corrcoef(window_ranges, df_scores_known_target_w["flussOracle"].tolist())[1,0]}')
print(f'Pearson Correlation without dilation, number of CPs: {np.corrcoef(number_of_cps,  df_scores_known_target_w["fluss1"].tolist())[1,0]}')
print(f'Pearson Correlation with dilation (Oracle), number of CPs: {np.corrcoef(number_of_cps, df_scores_known_target_w["flussOracle"].tolist())[1,0]}')
print(f'Pearson Correlation without dilation, ts length: {np.corrcoef(ts_lengths,  df_scores_known_target_w["fluss1"].tolist())[1,0]}')
print(f'Pearson Correlation with dilation (Oracle), ts length: {np.corrcoef(ts_lengths, df_scores_known_target_w["flussOracle"].tolist())[1,0]}')

# window coverages
scatter_fluss1 = plt.scatter(window_ranges, df_scores_known_target_w["fluss1"].tolist(), color='orange')
plt.xlabel('Window Coverage', fontsize=fontsize)
plt.ylabel('Covering Score', fontsize=fontsize)
legend_labels = [scatter_fluss1]
legend_texts = ['without dilation']
plt.legend(legend_labels, legend_texts)
# plt.savefig(f'./{experiment_name}window_range_d1.png', bbox_inches='tight')
plt.show()

scatter_flussOracle = plt.scatter(window_ranges, df_scores_known_target_w["flussOracle"].tolist(), color='blue')
plt.xlabel('Window Range', fontsize=fontsize)
plt.ylabel('Covering Score', fontsize=fontsize)
legend_labels = [scatter_flussOracle]
legend_texts = ['with dilation']
plt.legend(legend_labels, legend_texts)
# plt.savefig(f'./{experiment_name}window_range_dilation.png', bbox_inches='tight')
plt.show()

# number of CPs
scatter_fluss1 = plt.scatter(number_of_cps, df_scores_known_target_w["fluss1"].tolist(), color='orange')
plt.xlabel('Number of CPs', fontsize=fontsize)
plt.ylabel('Covering Score', fontsize=fontsize)
legend_labels = [scatter_fluss1]
legend_texts = ['without dilation']
plt.legend(legend_labels, legend_texts)
# plt.savefig(f'./{experiment_name}CPs_d1.png', bbox_inches='tight')
plt.show()

scatter_flussOracle = plt.scatter(number_of_cps, df_scores_known_target_w["flussOracle"].tolist(), color='blue')
plt.xlabel('Number of CPs', fontsize=fontsize)
plt.ylabel('Covering Score', fontsize=fontsize)
legend_labels = [scatter_flussOracle]
legend_texts = ['with dilation']
plt.legend(legend_labels, legend_texts)
# plt.savefig(f'./{experiment_name}CPs_dilation.png', bbox_inches='tight')
plt.show()

# ts length
scatter_fluss1 = plt.scatter(ts_lengths, df_scores_known_target_w["fluss1"].tolist(), color='orange')
plt.xlabel('Time Series Length', fontsize=fontsize)
plt.ylabel('Covering Score', fontsize=fontsize)
legend_labels = [scatter_fluss1]
legend_texts = ['without dilation']
plt.legend(legend_labels, legend_texts)
# plt.savefig(f'./{experiment_name}tslength_d1.png', bbox_inches='tight')
plt.show()

scatter_flussOracle = plt.scatter(ts_lengths, df_scores_known_target_w["flussOracle"].tolist(), color='blue')
plt.xlabel('Time Series Length', fontsize=fontsize)
plt.ylabel('Covering Score', fontsize=fontsize)
legend_labels = [scatter_flussOracle]
legend_texts = ['with dilation']
plt.legend(legend_labels, legend_texts)
# plt.savefig(f'./{experiment_name}tslength_dilation.png', bbox_inches='tight')
plt.show()
```
