```python
import matplotlib.pyplot as plt
import seaborn as sns

# data
window_ranges = [65,70,82,96,128,128,128,136,140,144,150,152,166,275,301,301,319]
f1_score_d1 = [0.0,0.2,0.356,0.0,0.0,0.258,0.338,0.780,0.0,0.342,0.0,0.292,0.0,0.4,0.578,0.2,0.292]
f1_score_dilation = [0.075,0.1,0.356,0.2,0.0,0.35,0.375,0.891,0.05,0.55,0.029,0.2,0.0,0.2,0.378,0.150,0.378]

# average performance
def mean(list):
    return sum(list) / len(list)
print(f'AV Performance d1: {mean(f1_score_d1)}')
print(f'AV Performance dilation: {mean(f1_score_dilation)}')


scatter1 = plt.scatter(window_ranges, f1_score_d1, alpha=0.7)
scatter2 = plt.scatter(window_ranges, f1_score_dilation, alpha=0.7)

plt.xlabel('Window Range')
plt.ylabel('F1 Score')
plt.title('Correlation between Window Range and F1-Score')

legend_labels = [scatter1, scatter2]
legend_texts = ['without dilation', 'with dilation']
plt.legend(legend_labels, legend_texts)

plt.show()
```
