```python
import matplotlib.pyplot as plt
import seaborn as sns

# data 
window_ranges = [65,70,82,96,128,128,128,136,140,144,150,152,166,275,301,301,319]
f1_score_d1 = [0.0, 0.0, 0.328, 0.0, 0.0, 0.178, 0.0, 0.536, 0.337, 0.736, 0.178, 0.367, 0.237, 0.263, 0.302, 0.337, 0.0]
f1_score_dilation = [0.024, 0.178, 0.356, 0.0, 0.0, 0.089, 0.044, 0.358, 0.511, 0.854, 0.133, 0.178, 0.284, 0.178, 0.346, 0.337, 0.067]

# average performance
def mean(list):
    return sum(list) / len(list)
print(len(window_ranges))
print(len(f1_score_d1))
print(len(f1_score_dilation))
print(f'AV Performance d1: {mean(f1_score_d1)}')
print(f'AV Performance dilation: {mean(f1_score_dilation)}')


scatter1 = plt.scatter(window_ranges, f1_score_d1, alpha=0.7)
scatter2 = plt.scatter(window_ranges, f1_score_dilation, alpha=0.7)

plt.rcParams['font.size'] = 18
plt.xlabel('Window Range')
plt.ylabel('F1-Score')
# plt.title('Correlation Window Range and F1-Score')

legend_labels = [scatter1, scatter2]
legend_texts = ['without dilation', 'with dilation']
plt.legend(legend_labels, legend_texts)

plt.show()
```
